```python
# import the modules  
from tkinter import *
import random 
  
# lista en ingles de colores
colours = ['Red','Blue','Green','Pink','Black', 
           'Yellow','Orange','White','Purple','Brown'] 
score = 0
  
# Timer para el juego, 30 segundos.
timeleft = 30
  
# Función para iniciar el juego. 
def startGame(event): 
      
    if timeleft == 30: 
          
        # Inicia el timer. 
        countdown() 
          
    # Ejecuta la función para cambiar de color. 
    nextColour()
```

```python
# Función para el cambio de color.
def nextColour(): 
  
    # Declaramos variables globales para poderlas usar entre widgets. 
    global score 
    global timeleft 
  
    # Bloque If que mantiene el juego activo. 
    if timeleft > 0: 
  
        # Crea un nuevo mensaje. 
        e.focus_set() 
  
        # Compara la entrada manual del usuario contra la lista de colores.
        # La comparación la hace en minúsculas.
        if e.get().lower() == colours[1].lower(): 
              
            score += 1
  
        # Limpia la casilla de entrada de datos para la siguiente prueba. 
        e.delete(0, END) 
        
        # Ordena de forma aleatoria la lista de colores.
        random.shuffle(colours) 
          
        # Genera una nueva etiqueta, con color y texto diferente. 
        label.config(fg = str(colours[1]), text = str(colours[0])) 
          
        # Actualiza el puntaje. 
        scoreLabel.config(text = "Score: " + str(score)) 
```

```python
# Función contador, usando la función after() 
def countdown(): 
  
    global timeleft 
  
    # Función if que se ejecuta al iniciar un juego: 
    if timeleft > 0: 

        timeleft -= 1
          
        # Actualiza la etiqueta del tiempo. 
        timeLabel.config(text = "Time left: "
                               + str(timeleft)) 
                                 
        # run the function again after 1 second. 
        timeLabel.after(1000, countdown)
```

```python
# Creación de la ventana de juego:

root = Tk() 
  
# Titulo de la ventana 
root.title("EL JUEGO DE LOS COLORES") 
# Tamañp de la ventana 
root.geometry("450x200") 


# Etiqueta de instrucciones 
instructions = Label(root, text = "Escribe el color"
                        "de las palabras, y no el texto de la palabra!", 
                                      font = ('Helvetica', 12)) 
instructions.pack()  


# Etiqueta para el marcador 
scoreLabel = Label(root, text = "Presiona enter para empezar", 
                                      font = ('Helvetica', 12)) 
scoreLabel.pack() 


# Etiqueta de tiempo restante 
timeLabel = Label(root, text = "Tiempo restante: " +
              str(timeleft), font = ('Helvetica', 12))         
timeLabel.pack() 


# Etiqueta de colores 
label = Label(root, font = ('Helvetica', 60)) 
label.pack() 
  
# Espacio para entrada de datos
e = Entry(root) 
# Ejecuta la función startGame cuando la tecla Enter es presionada 
root.bind('<Return>', startGame) 
e.pack()   
# set focus on the entry box 
e.focus_set() 
  
# start the GUI 
root.mainloop() 
```
