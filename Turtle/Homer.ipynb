```python
# Homero J Simpson

import turtle

t=turtle.Pen()
t.speed(0)

# Ojo, Nariz y Boca
t.circle(50)
t.up()
t.goto(45,30)
t.down()
t.goto(115,30)
t.left(180)
t.circle(25,-180)
t.goto(80,-20)
t.left(180)
t.circle(65,90)
x=t.position()

t.up()
t.goto(110,-20)
t.down()
t.left(60)
for i in range (60):
    if i<15:
        n=2
    else:
        n=1
    t.forward(n)
    t.right(1)
a=t.position()

# Ojo derecho
t.up()
t.goto(40,80)
t.down()
t.left(145)
for i in range (175):
    if i==80:
        ceja=t.position()
    t.forward(0.9)
    t.right(1)
    
# Labio superior
t.up()
t.goto(a)
t.down()
t.left(270)
t.circle(10,-120)
t.left(150)
t.forward(100)
t.left(90)

# Labio inferior
t.up()
t.goto(x)
t.down()
t.circle(65,90)
t.forward(5)
t.circle(40,90)
t.left(90)
t.forward(20)
t.right(180)
t.forward(22)
t.circle(20,75)


t.up()
t.goto(51.03,-95.69)
t.down()
t.circle(20,30)
t.up()
t.goto(51.03,-95.69)
t.down()
t.left(180)
t.circle(20,30)

# Ceja
t.up()
t.goto(ceja)
t.down()
t.left(105)
t.circle(15,110)
t.forward(30)
t.backward(15)

# Cabeza

t.right(60)
t.forward(60)
t.circle(100,180)
t.forward(150)

# Oreja
t.left(90)
t.circle(-20,90)
t.left(180)
t.circle(20,300)
t.up()
t.goto(-60,-45)
t.left(50)
t.down()
t.circle(12,180)
t.forward(5)
t.circle(9,280)

# Cuello

t.up()
t.goto(-70,-68)
t.left(50)
t.down()
t.left(25)
t.forward(120)
t.left(85)
t.forward(160)
t.left(100)
t.forward(40)

# M
t.up()
t.goto(-40,-15)
t.left(27.5)
t.down()
t.forward(50)
t.left(140)
t.forward(45)
t.right(140)
t.forward(50)
t.left(140)
t.forward(45)

# Cabello

t.up()
t.goto(-70,210)
t.left(202.5)
t.down()
t.circle(-40,190)

t.up()
t.goto(-90,190)
t.left(210)
t.down()
t.circle(-40,185)

# Iris
t.up()
t.goto(0,50)
t.down()
t.dot(15)
t.up()
t.goto(85,60)
t.down()
t.dot(15)
```
